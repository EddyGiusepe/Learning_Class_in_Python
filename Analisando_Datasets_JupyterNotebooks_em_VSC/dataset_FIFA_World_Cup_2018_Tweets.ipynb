```python
import polars as pl

import pandas as pd
import seaborn as sns

from summarytools import dfSummary
from skimpy import skim
```

## <font color="green">Biblioteca `skimpy`</font>

É uma ferramenta baseada em Jupyter que fornece um resumo de dados padronizado e abrangente. Isso inclui formato de dados, tipos de dados de coluna, estatísticas de resumo de coluna, gráficos de distribuição, estatísticas ausentes, etc., conforme mostrado abaixo:

```python
# df_fifa = pd.read_csv("./FIFA.csv")
# df_fifa.head()

df_pd = sns.load_dataset("iris")
df_pl = pl.from_pandas(df_pd)

df_pl
```

## <font color="green">Biblioteca `SummaryTools`</font>

<font color="orange">Que faz quase exatamente a mesma coisa que o Skimpy, ou seja, gera um relatório padronizado:</font>

```python
dfSummary(df_pd)
```

```python
skim(df_pd)
```

```python
skim(df_pl) # works with Polars DataFrame
```
